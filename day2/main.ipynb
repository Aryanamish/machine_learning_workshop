## <span style="color:#e8812c">Box Plot for identifying Outliers</span>

```python
ds.boxplot(column=['pregnancies'])
```

```python
pyplot.scatter(ds['bmi'], ds['bmi'])
```

## <span style="color:#e8812c">Fixing the Outliers by z-indexing method</span>

```python
t = 3
mean = round(ds['bmi'].mean(),2)
std = ds['bmi'].std()
for i in range(len(ds['bmi'])):
    z = (ds['bmi'][i]-mean)/std
    if z > t:
        ds['bmi'][i] = mean
```
